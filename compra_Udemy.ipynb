```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa
from matplotlib.ticker import FuncFormatter
```

```python
import pandas as pd
```

```python
excel = pd.read_csv(r'/home/ronald/Documentos/DatosPrueba/compra+curso.csv', sep=(";"))
```

```python
grafico = excel[["Compras_del_curso", "calificaciones"]]
```

```python
grafico.head()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots(figsize=(15,8))

ax.plot(grafico.Compras_del_curso, grafico.calificaciones, marker = "o", linestyle = "", markersize = 9, 
        color = "#F54927", markeredgecolor = "Black", markeredgewidth=1.5)

ax.set_xlabel("Compra del Curso", fontname ="serif", fontsize=15)
ax.set_ylabel("Calificaciones", fontname ="serif", fontsize=15)
ax.set_title("Ventas y Calificaciones", fontname ="serif", fontsize=15)

def formato(x,pos):
    return "%.0f mil compras"%(x/100)

ax.set_xticklabels(list(ax.get_xticks()), fontname ="serif", fontsize=10)
ax.xaxis.set_major_formatter(FuncFormatter(formato))

ax.set_yticklabels(list(ax.get_yticks()), fontname ="serif", fontsize=10)

plt.show()
```
