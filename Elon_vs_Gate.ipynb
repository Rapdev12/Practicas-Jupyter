```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as pa
from matplotlib.ticker import FuncFormatter
```

```python
import pandas as pd
```

```python
excel = pd.read_csv(r'/home/ronald/Documentos/DatosPrueba/finanzas1.csv', sep=(";"))
```

```python
excel.head()
```

```python
##Cambio de fecha
excel['Fecha'] = pd.to_datetime(excel['Fecha'], format= '%d/%m/%Y')
```

```python
excel.head()
```

```python
excel.dtypes
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(10,8))

ax.plot(excel['Elon Mosk'], label = 'Elon Mosk', lw = 2, color = 'Green',
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])

ax.plot(excel['Bill Gate'], label = 'Bill Gate', lw = 2, color = 'Orange',
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])
plt.show();
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(10,8))

ax.plot(excel['Elon Mosk'], label = 'Elon Mosk', lw = 2, color = 'Green',
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])

ax.plot(excel['Bill Gate'], label = 'Bill Gate', lw = 2, color = 'Orange',
       path_effects = [pa.SimpleLineShadow(), pa.Normal()])

ax.legend(shadow = True, prop = {"family":"Serif", "size":15})

ax.set_ylabel("USD", fontname ="Serif", fontsize=20)
ax.set_xlabel("Time", fontname ="Serif", fontsize=20)
ax.set_title("Elon Mosk vs Bill Gate", fontname ="Serif", fontsize=20)

plt.show();
```
